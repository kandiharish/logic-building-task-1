**Problem Statement 1:Smart Parking Lot Management System**

**Design a function to manage a smart parking lot.**
The system should:
Accept vehicle entry and exit logs


Calculate total parked vehicles


Identify peak parking usage


Alert if parking exceeds capacity




```python
def smart_parking_management(capacity, vehicle_logs):
    """
    Function to manage smart parking lot operations.

    Parameters:
    capacity (int): Maximum parking capacity
    vehicle_logs (list): List containing "IN" and "OUT" entries

    Returns:
    None (Prints current parked vehicles and parking status)
    """

    # Step 1: Initialize counters
    current_parked = 0   # Tracks current parked vehicles
    peak_parked = 0      # Tracks maximum vehicles parked at any time

    # Step 2: Process each log entry using loop
    for log in vehicle_logs:

        if log == "IN":
            current_parked += 1

            # Update peak usage if current exceeds previous peak
            if current_parked > peak_parked:
                peak_parked = current_parked

        elif log == "OUT":
            # Prevent negative parking count
            if current_parked > 0:
                current_parked -= 1

    # Step 3: Check capacity status using conditional logic
    if current_parked > capacity:
        status = "Alert: Parking Exceeded Capacity"
    else:
        status = "Available"

    # Step 4: Print results in expected format
    print(f"Currently Parked Vehicles: {current_parked}")
    print(f"Parking Status: {status}")

    # Optional: Peak usage insight (useful in real systems)
    # print(f"Peak Parking Usage: {peak_parked}")


# Example usage
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_management(parking_capacity, vehicle_logs)
```

**Problem Statement 2:Online Food Delivery Time Estimator**

**Create a function that estimates delivery time based on:**
Distance (km)


Weather condition


Traffic level


```python
def estimate_delivery_time(distance_km, traffic_level, weather_condition):
    """
    Function to estimate food delivery time based on distance,
    traffic level, and weather condition.

    Parameters:
    distance_km (int/float): Distance to delivery location in kilometers
    traffic_level (str): Traffic condition ("Low", "Medium", "High")
    weather_condition (str): Weather condition ("Clear", "Rainy", "Stormy")

    Returns:
    None (Prints estimated delivery time)
    """

    # Step 1: Calculate base delivery time
    # Assume average delivery speed gives 5 minutes per km
    estimated_time = distance_km * 5

    # Step 2: Add delay based on traffic level
    if traffic_level == "High":
        estimated_time += 10   # High traffic adds 10 minutes
    elif traffic_level == "Medium":
        estimated_time += 5    # Medium traffic adds 5 minutes
    # Low traffic adds no extra delay

    # Step 3: Add delay based on weather condition
    if weather_condition == "Rainy":
        estimated_time += 5    # Rain slows delivery slightly
    elif weather_condition == "Stormy":
        estimated_time += 15   # Storm causes major delay
    # Clear weather adds no delay

    # Step 4: Display final estimated delivery time
    print(f"Estimated Delivery Time: {estimated_time} minutes")


# Example usage
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

**Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

**Build a function that analyzes seat booking data and:**
Calculates occupancy percentage


Determines if show is Housefull


Suggests opening additional shows




```python
def analyze_theatre_occupancy(total_seats, booked_seats):
    """
    Function to analyze movie theatre seat occupancy.

    Parameters:
    total_seats (int): Total number of seats in the theatre
    booked_seats (list): List representing booked seats (each entry = 1)

    Returns:
    None (Prints occupancy percentage and show status)
    """

    # Step 1: Validate input
    if total_seats <= 0:
        print("Invalid total seats")
        return

    # Step 2: Count number of booked seats using list length
    booked_count = len(booked_seats)

    # Step 3: Calculate occupancy percentage
    occupancy_percentage = (booked_count / total_seats) * 100

    # Step 4: Determine show status using conditional logic
    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    # Step 5: Suggest opening additional shows if demand is high
    if occupancy_percentage >= 90:
        suggestion = "Suggestion: Open Additional Show"
    else:
        suggestion = "Suggestion: No Additional Show Needed"

    # Step 6: Print results in expected output format
    print(f"Occupancy: {int(occupancy_percentage)}%")
    print(f"Show Status: {status}")
    # Optional business insight
    # print(suggestion)


# Example usage
total_seats = 200
booked_seats = [1] * 150  # 150 booked seats

analyze_theatre_occupancy(total_seats, booked_seats)
```

**Problem Statement 4:Cloud Server Load Classification System**

**Create a function to classify server load based on CPU usage readings.**
Rules:
Average CPU < 50% → Normal


50%–80% → Warning


80% → Critical


```python
def classify_server_load(cpu_readings):
    """
    Function to calculate average CPU load and classify server status.

    Parameters:
    cpu_readings (list): List of CPU usage percentages

    Returns:
    None (Prints average CPU load and server status)
    """

    # Step 1: Validate input
    if len(cpu_readings) == 0:
        print("No CPU readings available")
        return

    # Step 2: Calculate total CPU usage using loop
    total_cpu = 0
    for reading in cpu_readings:
        total_cpu += reading

    # Step 3: Calculate average CPU usage
    average_cpu = total_cpu / len(cpu_readings)

    # Step 4: Classify server load using conditional logic
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    # Step 5: Display results in expected output format
    print(f"Average CPU Load: {int(average_cpu)}%")
    print(f"Server Status: {status}")


# Example usage
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

**Problem Statement 5:Smart Classroom Resource Usage Monitor**

**Design a function that tracks usage of classroom resources**(projector, AC, lights) and identifies overuse patterns.
Real-Time Use
Smart classrooms


Energy optimization systems


```python
def monitor_classroom_resources(resource_usage):
    """
    Function to track classroom resource usage and identify overuse.

    Parameters:
    resource_usage (dict): Dictionary with resource names as keys
                           and usage hours as values

    Returns:
    None (Prints overused resources and energy alert status)
    """

    # Step 1: Define overuse threshold (example: more than 8 hours)
    OVERUSE_THRESHOLD = 8

    # Step 2: Initialize list to store overused resources
    overused_resources = []

    # Step 3: Loop through dictionary to check usage
    for resource, hours in resource_usage.items():

        # Check if resource exceeds threshold
        if hours > OVERUSE_THRESHOLD:
            overused_resources.append(resource)

    # Step 4: Determine energy alert status
    if len(overused_resources) > 0:
        alert_status = "Yes"
    else:
        alert_status = "No"

    # Step 5: Print results in expected format
    if overused_resources:
        print(f"Overused Resources: {', '.join(overused_resources)}")
    else:
        print("Overused Resources: None")

    print(f"Energy Alert: {alert_status}")


# Example usage
resource_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_classroom_resources(resource_data)
```

**Problem Statement 6: Online Event Registration Capacity Controller**

**Create a function that manages event registrations by:**
Tracking registrations


Preventing overbooking


Triggering waitlist mode


```python
def manage_event_registration(event_capacity, total_registrations):
    """
    Function to manage event registrations, prevent overbooking,
    and trigger waitlist mode.

    Parameters:
    event_capacity (int): Maximum number of allowed registrations
    total_registrations (int): Total number of users attempting to register

    Returns:
    None (Prints confirmed registrations, waitlisted users, and status)
    """

    # Step 1: Initialize confirmed registration counter
    confirmed_registrations = 0

    # Step 2: Use loop to simulate registration process
    for i in range(total_registrations):

        # Allow registration only if capacity is available
        if confirmed_registrations < event_capacity:
            confirmed_registrations += 1
        else:
            # Stop once capacity is full
            break

    # Step 3: Calculate waitlisted users
    waitlisted_users = total_registrations - confirmed_registrations

    # Step 4: Determine registration status
    if confirmed_registrations == event_capacity:
        status = "Closed"
    else:
        status = "Open"

    # Step 5: Display results in expected output format
    print(f"Confirmed Registrations: {confirmed_registrations}")
    print(f"Waitlisted Users: {waitlisted_users}")
    print(f"Registration Status: {status}")


# Example usage
event_capacity = 100
registrations = 105

manage_event_registration(event_capacity, registrations)
```
