**Problem 1: Mobile Recharge Validation System**

**Problem Statement** : Create a function to validate a mobile recharge request.



```python
# Mobile Recharge Validation System

# List of valid recharge plans
VALID_PLANS = [199, 299, 399, 599]

# Function to validate recharge amount
def validate_recharge(amount):
    if amount < 50:
        print("‚ùå Recharge amount must be at least ‚Çπ50.")
        return False

    if amount not in VALID_PLANS:
        print("‚ùå Invalid recharge plan.")
        print("‚úÖ Available plans:", VALID_PLANS)
        return False

    return True


# Main program with retry option
def recharge_system():
    print("üì± Welcome to Mobile Recharge System")

    while True:
        try:
            amount = int(input("Enter recharge amount: ‚Çπ"))

            # Call validation function
            if validate_recharge(amount):
                print("‚úÖ Recharge successful for ‚Çπ", amount)
                break  # Exit loop if valid

            else:
                print("üîÅ Please try again.\n")

        except ValueError:
            print("‚ùå Invalid input. Please enter numbers only.\n")


# Run the system
recharge_system()
```

**Problem 2: Inventory Reorder Alert System**

**Problem Statement** :
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise

```python
# Warehouse Stock Monitoring System

# Dictionary of products and their stock quantities
products = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30,
    "Salt": 8
}

# Function to check stock status
def check_stock(product_dict):
    print("üì¶ Stock Status Report\n")

    for product, quantity in product_dict.items():

        if quantity < 15:
            print(f"‚ö†Ô∏è Reorder Alert: {product} stock is low ({quantity} units)")

        else:
            print(f"‚úÖ Stock OK: {product} has sufficient stock ({quantity} units)")


# Call the function
check_stock(products)
```

**Problem 3: Student Result Processing System**

**Problem Statement** :
Create a function that:
Accepts a list of student marks
Calculates average marks
Prints Pass if average ‚â• 50 else Fail

```python
# Student Result Processing System

# Function to calculate average and return result
def process_result(marks_list):

    # Calculate total using loop
    total = 0
    for mark in marks_list:
        total += mark

    # Calculate average
    average = total / len(marks_list)

    # Determine pass or fail
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result


# Example student marks
student_marks = [65, 45, 55, 70, 40]

# Call function
average_marks, final_result = process_result(student_marks)

# Display result
print("üìä Student Marks:", student_marks)
print("üìà Average Marks:", average_marks)
print("üéì Result:", final_result)
```

**Problem 4: Cab Fare Estimator with Retry Option**

**Problem Statement** :
Create a cab fare calculator:
Base fare = ‚Çπ50

‚Çπ12 per km

Peak hour ‚Üí 25% extra

Allow retry using while loop


```python
# Cab Fare Estimator with Retry Option

BASE_FARE = 50
RATE_PER_KM = 12
PEAK_MULTIPLIER = 1.25   # 25% extra


# Function to calculate fare
def calculate_fare(distance, is_peak):

    fare = BASE_FARE + (distance * RATE_PER_KM)

    if is_peak:
        fare = fare * PEAK_MULTIPLIER

    return fare


# Main system with retry option
def cab_fare_system():

    print("üöñ Welcome to Cab Fare Estimator\n")

    while True:
        try:
            distance = float(input("Enter distance in km: "))

            peak_input = input("Is it peak hour? (yes/no): ").lower()

            if peak_input == "yes":
                is_peak = True
            else:
                is_peak = False

            total_fare = calculate_fare(distance, is_peak)

            print(f"\nüí∞ Total Fare: ‚Çπ{total_fare:.2f}")


            retry = input("\nDo you want to calculate again? (yes/no): ").lower()

            if retry != "yes":
                print("‚úÖ Thank you for using Fare Estimator.")
                break

        except ValueError:
            print("‚ùå Invalid input. Please enter correct values.\n")


# Run system
cab_fare_system()
```

**Problem 5: Employee Attendance Eligibility Checker**

**Problem Statement**
Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ‚â• 75% else Not Eligible




```python
# Employee Attendance Eligibility Checker

# Function to calculate attendance percentage and eligibility
def check_attendance(attendance_list):

    present_count = 0

    # Count present days using loop
    for status in attendance_list:
        if status == "P":
            present_count += 1

    # Calculate percentage
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    # Check eligibility
    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"

    return percentage, result


# Example attendance record
attendance = ["P", "P", "A", "P", "P", "A", "P", "P", "P", "A"]

# Call function
attendance_percentage, eligibility = check_attendance(attendance)

# Display result
print("üìÖ Attendance Record:", attendance)
print(f"üìä Attendance Percentage: {attendance_percentage:.2f}%")
print("üèÜ Eligibility Status:", eligibility)
```

**Problem 6: Password Strength Checker**

**Problem Statement** :
Create a function to check password strength.
Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


```python
# Password Strength Checker

# Function to check password strength
def check_password_strength(password):

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check each character using loop
    for char in password:

        if char.isdigit():
            has_digit = True

        if char in special_chars:
            has_special = True

    # Check conditions
    if len(password) < 8:
        return "‚ùå Weak Password: Minimum length is 8"

    elif not has_digit:
        return "‚ùå Weak Password: Must contain at least one digit"

    elif not has_special:
        return "‚ùå Weak Password: Must contain at least one special character (@#$)"

    else:
        return "‚úÖ Strong Password"


# Example test
password = input("Enter password: ")

result = check_password_strength(password)

print(result)
```
